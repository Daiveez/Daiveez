```python
import matplotlib.pyplot as plt 
import numpy as np

arr1=np.array([80,85,90,95,100,105,110,115,120,125])
arr2=np.array([240,250,260,270,280,290,300,310,320,330])


font1=({'family':'serif','color':'blue','size':20})
font2=({'family':'serif','color':'darkred','size':10})

plt.plot(arr1,arr2,marker='*',ms=10,color='g')
plt.title("Average age",fontdict=font1,loc='left')
plt.xlabel("Average men age",fontdict=font2)
plt.ylabel("Average women age",fontdict=font2,)

#font dict is simply used to edit the graph 
#the loc is used to change the position of the title 

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y,marker='.',ms=10)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y,marker='*',ms=10)

plt.show()


#plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.

#plt.subplot(1, 2, 2)
#the figure has 1 row, 2 columns, and this plot is the second plot.
```
